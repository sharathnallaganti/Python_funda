```python
num = int(input(" Enter the number: "))
num_str = str(num)
power= len(num_str)
SumOfPower = 0

for digit in num_str:
    int_digit = int(digit)
    rasiedValue = int_digit ** power
    SumOfPower = SumOfPower + rasiedValue


if SumOfPower == num:
    print(num,"is the Armstrong number")
else:
     print(num,"is the not Armstrong number")
```

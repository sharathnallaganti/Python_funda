# Using String Slicing

```python
text = input("Enter the string to check: ")

if text == text[::-1]:
    print(text," is a palindrome!")

else:
    print(f"{text} is not a palindrome")
```

# Using the loop

```python
text = input("Enter the string: ")
is_palindrome  =  True

for i in range(len(text) //2):
    if text[i] != text[-(i+1)]:
        is_palindrome = False
        break

if is_palindrome:
    print(f"{text} is palindrome!")
else:
    print(f"{text} not is palindrome!")
    
```

# Palindrome Check for Numbers

```python
num = int(input("Enter the Number: "))

temp = num

rev = 0

while num >0:
    digit = num %10 
    rev = rev *10 + digit
    num = num//10

if temp == rev:
  print(f"{temp} is palindrome number")

else:
    print(f"{temp} is not palindrome number")
```

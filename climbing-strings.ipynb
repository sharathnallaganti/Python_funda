```python
class Solution:
    def climbStairs(self,no_steps):
        if no_steps <= 2:
            return no_steps

        prev2, prev1 = 1, 2
        for _ in range(3, no_steps + 1):
            current = prev1 + prev2
            prev2 = prev1
            prev1 = current

        return prev1




class Solution:
    def climbStairs(self,n):
        one , two  = 1, 1
        
        for i in range(n-1):
            temp =  one
            one = one + two
            two = temp
        
        return  one

            
 
```

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        n, m = len(haystack), len(needle)

        # (Optional) LeetCode sometimes allows empty needle: return 0 in that case
        if m == 0:
            return 0

        if m > n:
            return -1

        # Slide over all possible starting positions
        for i in range(n - m + 1):
            if haystack[i:i + m] == needle:
                return i

        return -1
```

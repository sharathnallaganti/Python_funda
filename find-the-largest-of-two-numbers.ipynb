#  Using loop

```python
nums =[1,7,9,-9,88,-999,76]
largest = nums[0]
for num in nums:
    if num>largest:
      largest = num

print("the largest number is ", largest)
```

# taking the input from the user

```python
n = int(input("how many numbers? : "))
numbers =[]

for i in range (n):
    num = float(input(f"enter the number{i+1}: "))
    numbers.append(num)

largest = numbers[0]
for num in numbers:
    if num> largest:
        largest = num
print("the largest number is :", largest)
```

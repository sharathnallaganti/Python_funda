```python
class Solution:
    def longestCommonPrefix(self, strs):
        if not strs:
            return ""

        prefix = strs[0]

        for s in strs[1:]:
            while not s.startswith(prefix):
                prefix = prefix[:-1]
                if prefix == "":
                    return ""
        return prefix
# Test the function
solution = Solution()

print(solution.longestCommonPrefix(["flower","flow","flight"]))  # Expected: "fl"
print(solution.longestCommonPrefix(["dog","racecar","car"]))    # Expected: ""
print(solution.longestCommonPrefix(["interspecies","interstellar","interstate"]))  # Expected: "inters"
```

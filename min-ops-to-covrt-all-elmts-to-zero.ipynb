```python
class Solution:
    def minOperations(self, nums: List[int]) -> int:
        ops = 0
        stack = []  # monotonic non-decreasing stack of heights

        for x in nums:
            # Remove heights that are larger than current x
            while stack and stack[-1] > x:
                stack.pop()

            # If we are starting a new positive plateau
            if x > 0 and (not stack or stack[-1] < x):
                stack.append(x)
                ops += 1
            # else: x == stack[-1] (same plateau) or x == 0 (stack already <= x)

        return ops
```

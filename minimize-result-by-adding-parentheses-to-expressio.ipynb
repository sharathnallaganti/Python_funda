```python
class Solution:
    def minimizeResult(self, expression: str) -> str:
        left, right = expression.split('+')
        best_val = float('inf')
        best_expr = ""

        # i = where '(' goes in left: left[:i] | left[i:]
        for i in range(len(left)):
            A = left[:i]
            B = left[i:]

            # j = where ')' goes in right: right[:j] | right[j:]
            for j in range(1, len(right) + 1):
                C = right[:j]
                D = right[j:]

                mul_left = int(A) if A else 1
                mul_right = int(D) if D else 1
                val = mul_left * (int(B) + int(C)) * mul_right

                if val < best_val:
                    best_val = val
                    best_expr = f"{A}({B}+{C}){D}"

        return best_expr
```

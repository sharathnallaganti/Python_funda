```python
class solution:
    def RemoveDuplicates(self,nums):
        if not nums:
            return 0
       
        k = 1  # next position for unique number
        
        for i in range(1, len(nums)):
            if nums[i] != nums[i - 1]:
                nums[k] = nums[i]
                k += 1
        
        return k
```

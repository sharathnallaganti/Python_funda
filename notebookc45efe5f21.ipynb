```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        i = len(s) - 1
        length = 0

        # 1) Skip trailing spaces
        while i >= 0 and s[i] == ' ':
            i -= 1

        # 2) Count characters of the last word
        while i >= 0 and s[i] != ' ':
            length += 1
            i -= 1

        return length
```

**Method call (belongs to an object)**

“This object knows how to do this action.

  object.method(arguments)


```python
s = "Hello World"

print(s.lower())              # string knows how to lowercase itself
print(s.replace(" ", "-"))    # string knows how to replace chars
print(s.split())              # string knows how to split itself

nums = [3, 1, 2]
nums.append(10)               # list knows how to append
nums.sort()                   # list knows how to sort itself
print(nums)
```

**Function or Class call (takes object as input)**

Function(object)
Class(object)

Meaning
“Use this object as input to produce something.”# 

```python
s = "hello"

print(len(s))         # len is a built-in function
print(int("123"))     # int converts string to integer
print(sorted([3,1,2]))  # sorted returns a new sorted list
```

Use functions/classes when you want:

a calculation (like len)

a conversion (like int, str)

a new object (like Counter, list, set, dict, sorted)

1. | Question                                      | Use                |
| --------------------------------------------- | ------------------ |
| Object changes itself?                        | `object.method()`  |
| Creating or transforming into something else? | `Function(object)` |


* | Task             | Correct           |
| ---------------- | ----------------- |
| Lowercase string | `name.lower()`    |
| Uppercase string | `name.upper()`    |
| Remove spaces    | `text.strip()`    |
| Split string     | `text.split()`    |
| Count char       | `text.count('a')` |

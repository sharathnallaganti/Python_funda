```python
from typing import List

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        
        # Traverse the digits from last index backward
        for i in range(len(digits)-1, -1, -1):
            if digits[i] < 9:        # No carry needed
                digits[i] += 1
                return digits
            digits[i] = 0            # Set 9 to 0 due to carry

        return [1] + digits          # If all digits were 9
```

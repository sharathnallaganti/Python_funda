```python
from typing import List

class Solution:
    def reconstructMatrix(self, upper: int, lower: int, colsum: List[int]) -> List[List[int]]:
        n = len(colsum)
        top = [0] * n
        bottom = [0] * n

        # First place the forced 2s
        for i, c in enumerate(colsum):
            if c == 2:
                top[i] = 1
                bottom[i] = 1
                upper -= 1
                lower -= 1
                if upper < 0 or lower < 0:
                    return []

        # Then place the 1s greedily
        for i, c in enumerate(colsum):
            if c == 1:
                if upper > 0:
                    top[i] = 1
                    upper -= 1
                else:
                    bottom[i] = 1
                    lower -= 1
                if upper < 0 or lower < 0:
                    return []

        return [top, bottom] if upper == 0 and lower == 0 else []
```

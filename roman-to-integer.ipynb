```python
class Solution:
    def romanToInt(self,s):
        # Step 1: Define the Roman numeral values
        roman_values = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }

        total = 0

        # Step 2: Loop through each character (except the last one)
        for i in range(len(s) - 1):
            current = roman_values[s[i]]
            next_val = roman_values[s[i + 1]]

            # Step 3: If the current value is smaller than the next one, subtract it
            if current < next_val:
                total -= current
            else:
                total += current

        # Step 4: Add the value of the last Roman numeral
        total += roman_values[s[-1]]
        
        return total

if __name__ == "__main__":
    sol = Solution()
    print(sol.romanToInt("III"))      # 3
    print(sol.romanToInt("LVIII"))    # 58
    print(sol.romanToInt("MCMXCIV"))  # 
```
